#Preprocessing

```python
import pandas as pd
data = pd.read_csv('train.txt', delimiter=' ')
print("Total  ", len(data.index))

print("Nbr of Negative", len(data[(data['class'] == 'negative')]))
```

```python
import csv
import glob 
pos_images = []
neg_images = []
list_patients = [] 
with open('train.txt') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=' ')
    line_count = 0
    pos = 0
    neg = 0
    for row in csv_reader:
        if line_count == 0:
          
            line_count += 1
        else:
           
            if (row[0] not in list_patients):
              #choose 1000 positive images
              if (row[2] == 'positive') and (row[1].endswith(".png")) and (pos <1000):
               
                pos_images.append('train/'+row[1])
                pos +=1
              #choose 1000 negative images
              if (row[2] == 'negative') and  (row[1].endswith(".png")) and (neg <1000):
                neg_images.append('train/'+row[1])
                neg +=1
              list_patients.append(row[0])
              
            line_count += 1
    
```

```python
import glob
import shutil
import os

src_dir = "train"
dst_dir = "gdrive/MyDrive/Datasets/covid/input/Yes/"

for filename in glob.iglob(os.path.join(src_dir, "*.png")):
  
  if(filename in pos_images):
    shutil.copy(filename, dst_dir)
```

```python
dst_dir = "gdrive/MyDrive/Datasets/covid/input/No/"

for filename in glob.iglob(os.path.join(src_dir, "*.png")):
  
  if(filename in neg_images):
    shutil.copy(filename, dst_dir)
```
